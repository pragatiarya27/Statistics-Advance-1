#Question 1: What is a random variable in probability theory?

- Answer:
In probability theory, a random variable is a variable whose possible values are determined by the outcome of a random experiment. It assigns numerical values to the results of random phenomena. Random variables help translate real-world uncertain events into a mathematical form that can be analyzed statistically.

For example, if you toss a fair coin, you can define a random variable X such that:

X=1 if the outcome is Heads
X=0 if the outcome is Tails

Thus, the random variable X represents the outcomes numerically, making it easier to calculate probabilities, expected values, and variances.

Mathematically, a random variable is a function that maps outcomes from a sample space (S) to real numbers (R):

#Question 2: What are the types of random variables?

- Answer:
There are two main types of random variables in probability theory:

1.Discrete Random Variable:

- Takes on a finite or countably infinite number of distinct values.

- Each value has an associated probability, and the sum of all probabilities equals 1.

- Example: The number of heads obtained in 3 coin tosses can be 0, 1, 2, or 3.

Probability Mass Function (PMF) describes the probability of each possible value.


$P(X=x_i) \ge 0, \sum P(X=x_i) = 1$





2.Continuous Random Variable:

- Can take on any value within a given interval (uncountably infinite).

- Probabilities are defined over ranges of values using a Probability Density Function (PDF).

- Example: The time taken for a computer to process a task (measured in seconds) can take any real value.
The probability that X lies between two values a and b is given by:


$$P(a \le X \le b) = \int_{a}^{b} f(x) dx$$where $f(x)$ is the probability density function.

#Question 3: Explain the difference between discrete and continuous distributions.

- Answer:
The main difference between discrete and continuous probability distributions lies in the type of random variable they describe and how probabilities are assigned.

| Feature | Discrete Distribution | Continuous Distribution |
|---|---|---|
| Type of Variable | Deals with discrete random variables that take countable values. | Deals with continuous random variables that take any value within a range. |
| Examples | Number of students in a class, number of heads in 5 coin tosses. | Height of students, time taken to complete a task. |
| Probability Assignment | Probability is assigned to each individual outcome using a Probability Mass Function (PMF). | Probability is assigned over intervals using a Probability Density Function (PDF). |
| Sum/Integral of Probabilities | The sum of all probabilities is 1: \(\sum P(X=x_i) = 1\) | The total area under the curve of the PDF is 1: integral to the value of -infinity to +infinity f(x)dx = 1\) |
| Probability of a Single Value | \(P(X=x)\) can be greater than 0. | \(P(X=x) = 0\); probability exists only over an interval. |

In summary, discrete distributions describe outcomes that can be counted (like rolling dice), while continuous distributions describe measurements that can take infinitely many values within a range (like temperature or time).

#Question 8: What is the concept of expected value in a probability distribution?

- Answer:
The expected value (EV) of a random variable is the long-term average or mean of its possible outcomes, weighted by their respective probabilities. It represents the value one can expect on average if an experiment is repeated many times.

For a discrete random variable, the expected value is:




Importance:

Helps in decision-making under uncertainty (e.g., in economics, insurance, and gambling).

Used to compute mean, variance, and standard deviation of random variables.

Forms the basis for many statistical and probabilistic models.

$$E(X) = \sum [x_i \cdot P(x_i)]$$



For a continuous random variable, it is defined as:




$$E(X) = \int_{-\infty}^{\infty} x f(x) dx$$where $f(x)$ is the probability density function.


Interpretation:

The expected value shows the center or balance point of the probability distribution.

It may not always be an actual possible outcome but represents the average result over the long run.


#Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

- Answer:






```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute mean and standard deviation
mean_value = np.mean(data)
std_value = np.std(data)

# Step 3: Display computed results
print("Mean of the distribution:", mean_value)
print("Standard Deviation of the distribution:", std_value)

# Step 4: Plot the histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Normal Distribution (Œº = 50, œÉ = 5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#(b) Python Code: Compute the mean sales and its 95% confidence interval

```python

import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)


confidence_level = 0.95
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_value * (std_sales / np.sqrt(n))

lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error


print("Mean Sales:", round(mean_sales, 2))
print("Standard Deviation:", round(std_sales, 2))
print(f"95% Confidence Interval: ({round(lower_bound, 2)}, {round(upper_bound, 2)})")
```
